```python
import numpy as m

sigmoid = lambda q: 1/(1+m.exp(-q))

xx = m.array([[0,0],[0,1],[1,0],[1,1]])
yy = m.array([[0.85],[0],[0],[1]])

w_1 = m.random.randn(2,4)
w_2 = m.random.randn(4,1)
biasA = m.random.randn(1,4)
biasB = m.random.randn(1,1)

for step in range(5000):
    layer1 = sigmoid(xx @ w_1 + biasA)
    final = sigmoid(layer1 @ w_2 + biasB)

print(final)

#ÙŽAi
```
